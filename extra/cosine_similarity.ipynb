```python
## DATA VIS: COSINE SIMILARITY PLOTS ##
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold, StratifiedKFold, train_test_split
import os
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.manifold import MDS
```

```python
splits = 10
```

```python
# paths to load full proteomics data
current_dir = os.getcwd()
data_dir = os.path.join(current_dir, 'split_data/10_fold/linear')

# move this into another script for ten-fold, maybe for 5-fold
X_train_list = []
X_test_list = []
for i in range(1, splits + 1):
# Reset X_train and X_test lists to drop the ID col, these were included before to save in the full files but can be dropped now
    train_file = os.path.join(data_dir, f'train_{i}.csv')
    test_file = os.path.join(data_dir, f'test_{i}.csv')
    train_df = pd.read_csv(train_file)
    test_df = pd.read_csv(train_file)
    X_train = train_df.drop(columns=['mtx_binary', 'Unnamed: 0', 'EAC_ID'])
    X_test = test_df.drop(columns=['mtx_binary', 'Unnamed: 0', 'EAC_ID'])
    # Append to respective lists
    X_train_list.append(X_train)
    X_test_list.append(X_test)
```

```python
# Calculate cosine similarity between the 10-fold training datasets
similarities = []
for i in range(splits):
    for j in range(splits):
        if i == j:
            similarities.append(1.0)
        else:
            similarity = cosine_similarity(X_train_list[i].mean(axis=0).values.reshape(1, -1),
                                           X_train_list[j].mean(axis=0).values.reshape(1, -1))[0][0]
            similarities.append(similarity)

similarity_matrix = np.array(similarities).reshape(splits, splits)

# Visualize the similarity matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(similarity_matrix, annot=True, cmap='coolwarm', vmin=0, vmax=1, 
            xticklabels=[f'Train {i+1}' for i in range(splits)], yticklabels=[f'Train {i+1}' for i in range(splits)])
plt.title('Cosine Similarity between K-Fold Training Datasets')
plt.show()
```
